```python
from google.cloud import bigquery
from google.colab import auth

# 1. Authenticate (Pop-up will ask for permission)
auth.authenticate_user()

# 2. Connect to BigQuery
project_id = 'gdelt-dataviz-project' # Replace with your Project ID (e.g., gdelt-dataviz-project)
client = bigquery.Client(project=project_id)

table_name = 'weekly_media_conflict'

# 3. Query directly
query = f"""
SELECT *
FROM `gdelt-dataviz-project.gdelt_agg.{table_name}`
ORDER BY mention_week DESC
"""

# 4. Save to Pandas DataFrame
df = client.query(query).to_dataframe()

# 5. Save to local CSV (in Colab environment)
df.to_csv('conflict_data_new.csv', index=False)
print("Data loaded! Rows:", len(df))
```
